```python
successful_rolls_init = []
current_roll_init = []

def enumerate_rolls(N,M,successful_rolls,current_roll):
        
    #fail condition if sum of dice less than actual number of dice - also works when total hits 0 and total negative
    if M<N: 
        return successful_rolls
    
    #even if we roll all sixes, can't reach sum - also works when total hits 0 and remaining sum too big
    if M>6*N: 
        return successful_rolls
    
    #following loop can only be reached if N=0 AND M=0, due to previous two conditions
    if N==0:
        return successful_rolls + [current_roll]
    
    for roll in range(1,7):
        successful_rolls = enumerate_rolls(N-1,M-roll, successful_rolls, current_roll + [roll])
    
    return successful_rolls #only gets reached when recursion branches exhausted
```

```python
import collections
import functools

class memoized(object):
   '''Decorator. Caches a function's return value each time it is called.
   If called later with the same arguments, the cached value is returned
   (not reevaluated).
   '''
   def __init__(self, func):
      self.func = func
      self.cache = {}
   def __call__(self, *args):
      if not isinstance(args, collections.Hashable):
         # uncacheable. a list, for instance.
         # better to not cache than blow up.
         return self.func(*args)
      if args in self.cache:
         return self.cache[args]
      else:
         value = self.func(*args)
         self.cache[args] = value
         return value
   def __repr__(self):
      '''Return the function's docstring.'''
      return self.func.__doc__
   def __get__(self, obj, objtype):
      '''Support instance methods.'''
      return functools.partial(self.__call__, obj)
```

```python
def enumerate_rolls_nu(N,M,successful_rolls,current_roll):

    if check_possible(N,M):

        #following loop can only be reached if N=0 AND M=0, due to output of check_possible
        if N==0:
            return successful_rolls + [current_roll]
        
        for roll in range(1,7):
            successful_rolls = enumerate_rolls_nu(N-1,M-roll, successful_rolls, current_roll + [roll])
            
    else: #if we have a combo of N
        return successful_rolls
    
    return successful_rolls #only gets reached when recursion branches exhausted
```

```python
#check_possible gets called over and over again with same N and M, therefore we use memoize class to speed up
@memoized
def check_possible(N,M):
    #fail condition if sum of dice less than actual number of dice - also works when total hits 0 and total negative
    if M<N: 
        return False
    
    #even if we roll all sixes, can't reach sum - also works when total hits 0 and remaining sum too big
    if M>6*N: 
        return False
    
    return True
```

```python
def enumerate_rolls_final(N,M,successful_rolls,current_roll):

    #this function only gets called in the first place if it's a viable roll, so we only need to check N==0
    if N==0:
        return successful_rolls + [current_roll]
        
    for roll in range(1,7):
        if check_possible(N-1,M-roll):
            successful_rolls = enumerate_rolls_final(N-1,M-roll, successful_rolls, current_roll + [roll])
    
    return successful_rolls #only gets reached when recursion branches exhausted
```
